```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import cv2
from lime import lime_image
from skimage.segmentation import mark_boundaries
import os
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')

print("TensorFlow version:", tf.__version__)
print("GPU Available:", tf.config.list_physical_devices('GPU'))


def load_and_preprocess_images(data_path, num_samples=50, img_size=(224, 224)):
    """Load a subset of images for LIME optimization"""
    images = []
    labels = []
    
    categories = ['glioma', 'meningioma', 'notumor', 'pituitary']
    
    for idx, category in enumerate(categories):
        category_path = os.path.join(data_path, category)
        if not os.path.exists(category_path):
            continue
            
        files = os.listdir(category_path)[:num_samples//4]
        
        for file in files:
            img_path = os.path.join(category_path, file)
            img = cv2.imread(img_path)
            if img is not None:
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img = cv2.resize(img, img_size)
                img = img / 255.0
                images.append(img)
                labels.append(idx)
    
    return np.array(images), np.array(labels)

# Load testing data (minimal 4 images - 1 per class)
test_path = '/kaggle/input/brain-tumor-mri-dataset/Testing'
X_test, y_test = load_and_preprocess_images(test_path, num_samples=4)
print(f"Loaded {len(X_test)} test images")

# PART 3: Load Pre-trained Models

def load_models():
    """Load the optimized models"""
    models = {}
    model_paths = {
        'GWO': '/kaggle/input/data-outputs/grey_wolf_optimized_best.kerass',
        'WOA': '/kaggle/input/data-outputs/whale_optimized_best.keras',
        'GWO_Alt': '/kaggle/input/data-outputs/whale_optimized.keras',
        'WOA_Alt': '/kaggle/input/data-outputs/wolf_optimized.keras'
    }
    
    for name, path in model_paths.items():
        if os.path.exists(path):
            try:
                models[name] = keras.models.load_model(path)
                print(f"Loaded model: {name}")
            except Exception as e:
                print(f"Error loading {name}: {e}")
    
    # Use the first available model
    if models:
        selected_model = list(models.values())[0]
        selected_name = list(models.keys())[0]
        print(f"\nUsing model: {selected_name}")
        return selected_model
    else:
        raise ValueError("No models found!")

model = load_models()

# PART 4: LIME Evaluation Metrics

def evaluate_lime_explanation(model, image, lime_params, num_runs=2):
    """
    Evaluate LIME explanation quality based on:
    1. Stability: Consistency across multiple runs
    2. Fidelity: How well explanation matches model behavior
    3. Complexity: Number of features used
    """
    
    num_features = lime_params['num_features']
    num_samples = lime_params['num_samples']
    kernel_width = lime_params['kernel_width']
    
    explainer = lime_image.LimeImageExplainer()
    
    # Run LIME multiple times to measure stability (reduced to 2)
    explanations = []
    masks = []
    
    for _ in range(num_runs):
        explanation = explainer.explain_instance(
            image,
            model.predict,
            top_labels=1,
            hide_color=0,
            num_samples=num_samples,
            segmentation_fn=lambda x: cv2.resize(
                cv2.resize(x, (14, 14), interpolation=cv2.INTER_LINEAR),
                (x.shape[1], x.shape[0]),
                interpolation=cv2.INTER_NEAREST
            )
        )
        
        # Get the top label
        top_label = explanation.top_labels[0]
        
        # Get explanation mask
        temp, mask = explanation.get_image_and_mask(
            top_label,
            positive_only=True,
            num_features=num_features,
            hide_rest=False
        )
        
        explanations.append(explanation)
        masks.append(mask)
    
    # Calculate stability (IoU between masks)
    stability_scores = []
    for i in range(len(masks)-1):
        intersection = np.sum((masks[i] > 0) & (masks[i+1] > 0))
        union = np.sum((masks[i] > 0) | (masks[i+1] > 0))
        iou = intersection / (union + 1e-10)
        stability_scores.append(iou)
    
    stability = np.mean(stability_scores) if stability_scores else 0
    
    # Calculate fidelity (prediction consistency on perturbed images)
    original_pred = model.predict(np.expand_dims(image, axis=0), verbose=0)
    original_class = np.argmax(original_pred)
    
    # Create masked version using most important features
    avg_mask = np.mean(masks, axis=0)
    important_regions = avg_mask > 0
    
    masked_image = image.copy()
    masked_image[~important_regions] = 0
    
    masked_pred = model.predict(np.expand_dims(masked_image, axis=0), verbose=0)
    fidelity = np.max(masked_pred)
    
    # Complexity score (prefer simpler explanations)
    complexity = 1.0 / (num_features + 1)
    
    # Combined score (weighted average)
    score = 0.4 * stability + 0.4 * fidelity + 0.2 * complexity
    
    return score, stability, fidelity, complexity

# ============================================================================
# PART 5: Grey Wolf Optimizer (GWO)
# ============================================================================

class GreyWolfOptimizer:
    """Grey Wolf Optimizer for LIME parameter optimization"""
    
    def __init__(self, obj_function, lb, ub, dim, n_wolves=10, max_iter=20):
        self.obj_function = obj_function
        self.lb = np.array(lb)
        self.ub = np.array(ub)
        self.dim = dim
        self.n_wolves = n_wolves
        self.max_iter = max_iter
        
        # Initialize wolf positions
        self.positions = np.random.uniform(
            self.lb, self.ub, (self.n_wolves, self.dim)
        )
        
        # Alpha, Beta, Delta wolves (best 3)
        self.alpha_pos = np.zeros(self.dim)
        self.alpha_score = float('inf')
        
        self.beta_pos = np.zeros(self.dim)
        self.beta_score = float('inf')
        
        self.delta_pos = np.zeros(self.dim)
        self.delta_score = float('inf')
        
        self.convergence_curve = []
    
    def optimize(self):
        """Run the optimization"""
        print("\n=== Grey Wolf Optimizer ===")
        
        for iteration in range(self.max_iter):
            # Update alpha, beta, delta
            for i in range(self.n_wolves):
                fitness = self.obj_function(self.positions[i])
                
                if fitness < self.alpha_score:
                    self.delta_score = self.beta_score
                    self.delta_pos = self.beta_pos.copy()
                    self.beta_score = self.alpha_score
                    self.beta_pos = self.alpha_pos.copy()
                    self.alpha_score = fitness
                    self.alpha_pos = self.positions[i].copy()
                elif fitness < self.beta_score:
                    self.delta_score = self.beta_score
                    self.delta_pos = self.beta_pos.copy()
                    self.beta_score = fitness
                    self.beta_pos = self.positions[i].copy()
                elif fitness < self.delta_score:
                    self.delta_score = fitness
                    self.delta_pos = self.positions[i].copy()
            
            # Update parameter a (linearly decreases from 2 to 0)
            a = 2 - iteration * (2 / self.max_iter)
            
            # Update positions
            for i in range(self.n_wolves):
                for j in range(self.dim):
                    r1, r2 = np.random.random(2)
                    A1 = 2 * a * r1 - a
                    C1 = 2 * r2
                    
                    D_alpha = abs(C1 * self.alpha_pos[j] - self.positions[i, j])
                    X1 = self.alpha_pos[j] - A1 * D_alpha
                    
                    r1, r2 = np.random.random(2)
                    A2 = 2 * a * r1 - a
                    C2 = 2 * r2
                    
                    D_beta = abs(C2 * self.beta_pos[j] - self.positions[i, j])
                    X2 = self.beta_pos[j] - A2 * D_beta
                    
                    r1, r2 = np.random.random(2)
                    A3 = 2 * a * r1 - a
                    C3 = 2 * r2
                    
                    D_delta = abs(C3 * self.delta_pos[j] - self.positions[i, j])
                    X3 = self.delta_pos[j] - A3 * D_delta
                    
                    self.positions[i, j] = (X1 + X2 + X3) / 3
                
                # Boundary check
                self.positions[i] = np.clip(self.positions[i], self.lb, self.ub)
            
            self.convergence_curve.append(self.alpha_score)
            print(f"Iteration {iteration+1}/{self.max_iter}, Best Score: {-self.alpha_score:.4f}")
        
        return self.alpha_pos, -self.alpha_score, self.convergence_curve

# PART 6: Whale Optimization Algorithm (WOA)

class WhaleOptimizer:
    """Whale Optimization Algorithm for LIME parameter optimization"""
    
    def __init__(self, obj_function, lb, ub, dim, n_whales=10, max_iter=20):
        self.obj_function = obj_function
        self.lb = np.array(lb)
        self.ub = np.array(ub)
        self.dim = dim
        self.n_whales = n_whales
        self.max_iter = max_iter
        
        # Initialize whale positions
        self.positions = np.random.uniform(
            self.lb, self.ub, (self.n_whales, self.dim)
        )
        
        # Best position (prey)
        self.best_pos = np.zeros(self.dim)
        self.best_score = float('inf')
        
        self.convergence_curve = []
    
    def optimize(self):
        """Run the optimization"""
        print("\n=== Whale Optimization Algorithm ===")
        
        for iteration in range(self.max_iter):
            # Update best solution
            for i in range(self.n_whales):
                fitness = self.obj_function(self.positions[i])
                
                if fitness < self.best_score:
                    self.best_score = fitness
                    self.best_pos = self.positions[i].copy()
            
            # Update parameter a (linearly decreases from 2 to 0)
            a = 2 - iteration * (2 / self.max_iter)
            a2 = -1 + iteration * (-1 / self.max_iter)
            
            # Update positions
            for i in range(self.n_whales):
                r = np.random.random()
                A = 2 * a * r - a
                C = 2 * r
                
                b = 1  # Defines the shape of logarithmic spiral
                l = np.random.uniform(-1, 1)
                
                p = np.random.random()
                
                for j in range(self.dim):
                    if p < 0.5:
                        if abs(A) < 1:
                            # Encircling prey
                            D = abs(C * self.best_pos[j] - self.positions[i, j])
                            self.positions[i, j] = self.best_pos[j] - A * D
                        else:
                            # Search for prey (exploration)
                            rand_whale_idx = np.random.randint(0, self.n_whales)
                            X_rand = self.positions[rand_whale_idx]
                            D = abs(C * X_rand[j] - self.positions[i, j])
                            self.positions[i, j] = X_rand[j] - A * D
                    else:
                        # Spiral updating position
                        D_prime = abs(self.best_pos[j] - self.positions[i, j])
                        self.positions[i, j] = D_prime * np.exp(b * l) * np.cos(2 * np.pi * l) + self.best_pos[j]
                
                # Boundary check
                self.positions[i] = np.clip(self.positions[i], self.lb, self.ub)
            
            self.convergence_curve.append(self.best_score)
            print(f"Iteration {iteration+1}/{self.max_iter}, Best Score: {-self.best_score:.4f}")
        
        return self.best_pos, -self.best_score, self.convergence_curve

# PART 7: Objective Function for LIME Optimization

def lime_objective_function(params):
    """
    Objective function to minimize (negative score to maximize)
    params: [num_features, num_samples, kernel_width]
    """
    num_features = int(params[0])
    num_samples = int(params[1])
    kernel_width = params[2]
    
    lime_params = {
        'num_features': num_features,
        'num_samples': num_samples,
        'kernel_width': kernel_width
    }
    
    # Evaluate on single image only for speed
    scores = []
    num_eval_images = 1
    
    for i in range(num_eval_images):
        try:
            score, _, _, _ = evaluate_lime_explanation(
                model, X_test[i], lime_params, num_runs=2
            )
            scores.append(score)
        except Exception as e:
            scores.append(0.0)
    
    avg_score = np.mean(scores)
    return -avg_score  # Negative because we minimize

# ============================================================================
# PART 8: Run Optimizations
# ============================================================================

# Define search space (minimal range for ultra-fast optimization)
# [num_features (5-12), num_samples (50-150), kernel_width (0.1-1.0)]
lb = [5, 50, 0.1]
ub = [12, 150, 1.0]
dim = 3

print("\n" + "="*60)
print("Starting LIME Parameter Optimization")
print("="*60)

# Grey Wolf Optimizer (ultra-minimal: 3 wolves, 2 iterations)
gwo = GreyWolfOptimizer(
    obj_function=lime_objective_function,
    lb=lb,
    ub=ub,
    dim=dim,
    n_wolves=3,
    max_iter=2
)

gwo_best_params, gwo_best_score, gwo_curve = gwo.optimize()

print("\n" + "="*60)
print("GWO Results:")
print(f"Best Parameters: num_features={int(gwo_best_params[0])}, "
      f"num_samples={int(gwo_best_params[1])}, "
      f"kernel_width={gwo_best_params[2]:.3f}")
print(f"Best Score: {gwo_best_score:.4f}")
print("="*60)

# Whale Optimization Algorithm (ultra-minimal: 3 whales, 2 iterations)
woa = WhaleOptimizer(
    obj_function=lime_objective_function,
    lb=lb,
    ub=ub,
    dim=dim,
    n_whales=3,
    max_iter=2
)

woa_best_params, woa_best_score, woa_curve = woa.optimize()

print("\n" + "="*60)
print("WOA Results:")
print(f"Best Parameters: num_features={int(woa_best_params[0])}, "
      f"num_samples={int(woa_best_params[1])}, "
      f"kernel_width={woa_best_params[2]:.3f}")
print(f"Best Score: {woa_best_score:.4f}")
print("="*60)

# PART 9: Visualization

# Plot convergence curves
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot([-x for x in gwo_curve], 'b-', linewidth=2, label='GWO')
plt.plot([-x for x in woa_curve], 'r-', linewidth=2, label='WOA')
plt.xlabel('Iteration', fontsize=12)
plt.ylabel('Best Score', fontsize=12)
plt.title('Convergence Curves', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

# Compare final scores
plt.subplot(1, 2, 2)
optimizers = ['GWO', 'WOA']
scores = [gwo_best_score, woa_best_score]
colors = ['blue', 'red']
plt.bar(optimizers, scores, color=colors, alpha=0.7)
plt.ylabel('Best Score', fontsize=12)
plt.title('Final Optimization Scores', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('/kaggle/working/optimization_comparison.png', dpi=150, bbox_inches='tight')
plt.show()

print("\nConvergence plot saved to: /kaggle/working/optimization_comparison.png")

# PART 10: Compare LIME Explanations

def visualize_lime_comparison(image, model, default_params, gwo_params, woa_params):
    """Compare LIME explanations with different parameter sets"""
    
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    
    param_sets = [
        ('Default', default_params),
        ('GWO Optimized', gwo_params),
        ('WOA Optimized', woa_params)
    ]
    
    for idx, (name, params) in enumerate(param_sets):
        explainer = lime_image.LimeImageExplainer()
        
        explanation = explainer.explain_instance(
            image,
            model.predict,
            top_labels=1,
            hide_color=0,
            num_samples=params['num_samples'],
            segmentation_fn=lambda x: cv2.resize(
                cv2.resize(x, (14, 14), interpolation=cv2.INTER_LINEAR),
                (x.shape[1], x.shape[0]),
                interpolation=cv2.INTER_NEAREST
            )
        )
        
        top_label = explanation.top_labels[0]
        temp, mask = explanation.get_image_and_mask(
            top_label,
            positive_only=True,
            num_features=params['num_features'],
            hide_rest=False
        )
        
        # Original image
        axes[0, idx].imshow(image)
        axes[0, idx].set_title(f'{name}\nOriginal Image', fontweight='bold')
        axes[0, idx].axis('off')
        
        # Explanation overlay - create a simple overlay instead of mark_boundaries
        explanation_img = image.copy()
        # Highlight important regions
        mask_binary = (mask > 0).astype(np.uint8)
        overlay = np.zeros_like(image)
        overlay[mask_binary == 1] = [0, 255, 0]  # Green highlight
        explanation_img = cv2.addWeighted(image, 0.7, overlay, 0.3, 0)
        
        axes[1, idx].imshow(explanation_img)
        pred = model.predict(np.expand_dims(image, axis=0), verbose=0)
        pred_class = np.argmax(pred)
        confidence = np.max(pred)
        
        title = f'{name}\nPrediction: Class {pred_class}\n'
        title += f'Confidence: {confidence:.2%}\n'
        title += f'Features: {params["num_features"]}, Samples: {params["num_samples"]}'
        
        axes[1, idx].set_title(title, fontweight='bold', fontsize=9)
        axes[1, idx].axis('off')
    
    plt.tight_layout()
    return fig

# Create comparison for a sample image
sample_idx = 0
default_params = {'num_features': 8, 'num_samples': 100, 'kernel_width': 0.25}
gwo_params = {
    'num_features': int(gwo_best_params[0]),
    'num_samples': int(gwo_best_params[1]),
    'kernel_width': gwo_best_params[2]
}
woa_params = {
    'num_features': int(woa_best_params[0]),
    'num_samples': int(woa_best_params[1]),
    'kernel_width': woa_best_params[2]
}

print("\nComparison visualization saved to: /kaggle/working/lime_comparison.png")

# PART 11: Detailed Evaluation Report

print("\n" + "="*60)
print("DETAILED EVALUATION REPORT")
print("="*60)

eval_images = X_test[:1]

for img_idx, image in enumerate(eval_images):
    print(f"\n--- Image {img_idx + 1} ---")
    
    for name, params in [('Default', default_params), 
                         ('GWO', gwo_params), 
                         ('WOA', woa_params)]:
        score, stability, fidelity, complexity = evaluate_lime_explanation(
            model, image, params, num_runs=2
        )
        
        print(f"\n{name} Parameters:")
        print(f"  Overall Score: {score:.4f}")
        print(f"  Stability: {stability:.4f}")
        print(f"  Fidelity: {fidelity:.4f}")
        print(f"  Complexity: {complexity:.4f}")

# PART 12: Save Optimized Parameters

results = {
    'GWO': {
        'params': {
            'num_features': int(gwo_best_params[0]),
            'num_samples': int(gwo_best_params[1]),
            'kernel_width': float(gwo_best_params[2])
        },
        'score': float(gwo_best_score)
    },
    'WOA': {
        'params': {
            'num_features': int(woa_best_params[0]),
            'num_samples': int(woa_best_params[1]),
            'kernel_width': float(woa_best_params[2])
        },
        'score': float(woa_best_score)
    }
}

import json
with open('/kaggle/working/optimized_lime_params.json', 'w') as f:
    json.dump(results, f, indent=4)

print("\n" + "="*60)
print("Optimized parameters saved to: /kaggle/working/optimized_lime_params.json")
print("="*60)

print("\n Optimization Complete!")
print("\nBest Configuration:")
if gwo_best_score > woa_best_score:
    print("GWO achieved the best results!")
    print(f"Parameters: {results['GWO']['params']}")
    print(f"Score: {gwo_best_score:.4f}")
else:
    print("WOA achieved the best results!")
    print(f"Parameters: {results['WOA']['params']}")
    print(f"Score: {woa_best_score:.4f}")
```
