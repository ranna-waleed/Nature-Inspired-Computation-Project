```python
import os
import gc
import numpy as np
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import time
import matplotlib.pyplot as plt
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# ============================================
# DATASET LOADING
# ============================================

def load_brain_tumor_dataset():
    """Load Brain Tumor MRI Dataset with optimized settings"""
    print("LOADING BRAIN TUMOR MRI DATASET")
    print("="*60)
    
    # Set dataset paths
    dataset_path = '/kaggle/input/brain-tumor-mri-dataset'
    train_dir = '//kaggle/input/d/masoudnickparvar/brain-tumor-mri-dataset/Training'
    test_dir = '/kaggle/input/d/masoudnickparvar/brain-tumor-mri-dataset/Testing'
    
    # Check if directories exist
    print(f"Training directory exists: {os.path.exists(train_dir)}")
    print(f"Testing directory exists: {os.path.exists(test_dir)}")
    
    if not os.path.exists(train_dir) or not os.path.exists(test_dir):
        raise FileNotFoundError("Dataset directories not found. Please check the paths.")
    
    # List classes
    classes = sorted(os.listdir(train_dir))
    print(f"\nClasses found: {classes}")
    
    # Count images
    print("\nImages per class (Training):")
    total_train = 0
    for cls in classes:
        class_path = os.path.join(train_dir, cls)
        num_images = len(os.listdir(class_path))
        total_train += num_images
        print(f"  {cls}: {num_images} images")
    print(f"  Total training images: {total_train}")
    
    # Data augmentation for training
    train_datagen = ImageDataGenerator(
        rescale=1./255,
        rotation_range=15,
        width_shift_range=0.1,
        height_shift_range=0.1,
        zoom_range=0.1,
        horizontal_flip=True,
        fill_mode='nearest',
        validation_split=0.15
    )
    
    # Only rescaling for validation
    val_datagen = ImageDataGenerator(
        rescale=1./255,
        validation_split=0.15
    )
    
    # Test data generator
    test_datagen = ImageDataGenerator(rescale=1./255)
    
    # Create generators with optimized image size
    print("\nCreating data generators...")
    
    train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(64, 64),
        batch_size=16,
        class_mode='categorical',
        subset='training',
        seed=42,
        shuffle=True
    )
    
    validation_generator = val_datagen.flow_from_directory(
        train_dir,
        target_size=(64, 64),
        batch_size=16,
        class_mode='categorical',
        subset='validation',
        seed=42,
        shuffle=False
    )
    
    test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=(64, 64),
        batch_size=16,
        class_mode='categorical',
        shuffle=False
    )
    
    print(f"\nDataset Summary:")
    print(f"Training samples: {train_generator.samples}")
    print(f"Validation samples: {validation_generator.samples}")
    print(f"Testing samples: {test_generator.samples}")
    print(f"Number of classes: {train_generator.num_classes}")
    print(f"Class indices: {train_generator.class_indices}")
    
    return train_generator, validation_generator, test_generator

def load_baseline_model():
    """Load your pre-trained baseline model"""
    baseline_path = '/kaggle/input/baseline-model/keras/default/1/baseline_model (1).keras'
    
    try:
        print(f"Attempting to load baseline model from: {baseline_path}")
        model = keras.models.load_model(baseline_path)
        print("Baseline model loaded successfully")
        
        # Check model input shape
        input_shape = model.input_shape
        print(f"   Model input shape: {input_shape}")
        
        # If model expects different input shape, adjust it
        if input_shape[1:] != (64, 64, 3):
            print(f"Model expects {input_shape[1:]}, adjusting input layer...")
            new_input = keras.layers.Input(shape=(64, 64, 3))
            
            x = new_input
            for layer in model.layers[1:]:
                layer.trainable = False
                x = layer(x)
            
            model = keras.Model(inputs=new_input, outputs=x)
        
        # Recompile
        model.compile(
            optimizer=keras.optimizers.Adam(learning_rate=0.001),
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        
        model.summary()
        return model
        
    except Exception as e:
        print(f"Error loading baseline model: {str(e)[:100]}")
        print("Creating a new baseline model")
        return create_baseline_model()

def create_baseline_model():
    """Create a baseline model"""
    print("Creating baseline model...")
    
    model = keras.Sequential([
        keras.layers.Conv2D(16, (3, 3), activation='relu', 
                           input_shape=(64, 64, 3), padding='same'),
        keras.layers.BatchNormalization(),
        keras.layers.MaxPooling2D((2, 2)),
        keras.layers.Dropout(0.2),
        
        keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        keras.layers.BatchNormalization(),
        keras.layers.MaxPooling2D((2, 2)),
        keras.layers.Dropout(0.2),
        
        keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        keras.layers.BatchNormalization(),
        keras.layers.MaxPooling2D((2, 2)),
        keras.layers.Dropout(0.3),
        
        keras.layers.GlobalAveragePooling2D(),
        
        keras.layers.Dense(128, activation='relu'),
        keras.layers.BatchNormalization(),
        keras.layers.Dropout(0.4),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dropout(0.2),
        keras.layers.Dense(4, activation='softmax')
    ])
    
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    print("Baseline model created")
    model.summary()
    
    return model

# ============================================
# WHALE OPTIMIZATION
# ============================================

class WhaleOptimization:
    def __init__(self, n_whales=4, max_iter=8):
        """Whale Optimization Algorithm"""
        self.n_whales = n_whales
        self.bounds = self.get_bounds()
        self.max_iter = max_iter
        self.best_solution = None
        self.best_fitness = float('inf')
        self.convergence_curve = []
    
    def get_bounds(self):
        """Hyperparameter bounds"""
        return {
            'learning_rate': (1e-4, 1e-2),
            'batch_size': (8, 24),
            'n_filters1': (8, 32),
            'n_filters2': (16, 48),
            'n_filters3': (32, 64),
            'dense_units': (64, 192),
            'dropout_rate': (0.1, 0.4)
        }
    
    def initialize_whales(self):
        """Initialize whale population"""
        whales = []
        for _ in range(self.n_whales):
            whale = {}
            for param, (min_val, max_val) in self.bounds.items():
                if param == 'batch_size':
                    whale[param] = int(np.random.uniform(min_val, max_val))
                else:
                    whale[param] = np.random.uniform(min_val, max_val)
            whales.append(whale)
        return whales
    
    def evaluate_whale(self, whale_params, train_gen, val_gen):
        """Train CNN with whale parameters"""
        try:
            tf.keras.backend.clear_session()
            gc.collect()
            
            model = keras.Sequential([
                keras.layers.Conv2D(int(whale_params['n_filters1']), (3, 3), 
                                   activation='relu', input_shape=(64, 64, 3), padding='same'),
                keras.layers.MaxPooling2D((2, 2)),
                keras.layers.Dropout(whale_params['dropout_rate']/3),
                
                keras.layers.Conv2D(int(whale_params['n_filters2']), (3, 3), 
                                   activation='relu', padding='same'),
                keras.layers.MaxPooling2D((2, 2)),
                keras.layers.Dropout(whale_params['dropout_rate']/2),
                
                keras.layers.Conv2D(int(whale_params['n_filters3']), (3, 3), 
                                   activation='relu', padding='same'),
                keras.layers.MaxPooling2D((2, 2)),
                keras.layers.Dropout(whale_params['dropout_rate']),
                
                keras.layers.GlobalAveragePooling2D(),
                
                keras.layers.Dense(int(whale_params['dense_units']), activation='relu'),
                keras.layers.Dropout(whale_params['dropout_rate']),
                keras.layers.Dense(train_gen.num_classes, activation='softmax')
            ])
            
            optimizer = keras.optimizers.Adam(learning_rate=whale_params['learning_rate'])
            model.compile(
                optimizer=optimizer,
                loss='categorical_crossentropy',
                metrics=['accuracy']
            )
            
            batch_size = min(16, int(whale_params['batch_size']))
            steps = min(5, train_gen.samples // batch_size)
            
            history = model.fit(
                train_gen,
                steps_per_epoch=steps,
                epochs=2,
                validation_data=val_gen,
                validation_steps=min(3, val_gen.samples // batch_size),
                verbose=0
            )
            
            val_loss = history.history['val_loss'][-1]
            return val_loss
            
        except Exception as e:
            print(f"  Evaluation error: {str(e)[:80]}...")
            return float('inf')
    
    def optimize(self, train_gen, val_gen):
        """Main WOA optimization loop"""
        print("\n" + "="*60)
        print("WHALE OPTIMIZATION ALGORITHM")
        print("="*60)
        print(f"Population: {self.n_whales} whales")
        print(f"Iterations: {self.max_iter}")
        print(f"Hyperparameters: {list(self.bounds.keys())}")
        print("-"*60)
        
        whales = self.initialize_whales()
        fitness = [self.evaluate_whale(w, train_gen, val_gen) for w in whales]
        
        best_idx = np.argmin(fitness)
        self.best_solution = whales[best_idx].copy()
        self.best_fitness = fitness[best_idx]
        
        print(f"Initial best loss: {self.best_fitness:.4f}")
        
        for iteration in tqdm(range(self.max_iter), desc="WOA Progress"):
            a = 2 - iteration * (2 / self.max_iter)
            
            for i in range(self.n_whales):
                r1, r2 = np.random.rand(), np.random.rand()
                A = 2 * a * r1 - a
                C = 2 * r2
                
                current_whale = whales[i].copy()
                p = np.random.rand()
                
                for param in self.bounds.keys():
                    if p < 0.5:
                        if abs(A) < 1:
                            D = abs(C * self.best_solution[param] - current_whale[param])
                            new_val = self.best_solution[param] - A * D
                        else:
                            random_idx = np.random.randint(0, self.n_whales)
                            random_whale = whales[random_idx]
                            D = abs(C * random_whale[param] - current_whale[param])
                            new_val = random_whale[param] - A * D
                    else:
                        distance = abs(self.best_solution[param] - current_whale[param])
                        l = (a - 1) * np.random.rand() + 1
                        new_val = distance * np.exp(l) * np.cos(2 * np.pi * l) + self.best_solution[param]
                    
                    min_val, max_val = self.bounds[param]
                    if param == 'batch_size':
                        new_val = int(np.clip(new_val, min_val, max_val))
                    else:
                        new_val = np.clip(new_val, min_val, max_val)
                    
                    current_whale[param] = new_val
                
                new_fitness = self.evaluate_whale(current_whale, train_gen, val_gen)
                
                if new_fitness < fitness[i]:
                    whales[i] = current_whale.copy()
                    fitness[i] = new_fitness
                    
                    if new_fitness < self.best_fitness:
                        self.best_solution = current_whale.copy()
                        self.best_fitness = new_fitness
            
            self.convergence_curve.append(self.best_fitness)
            
            if (iteration + 1) % 5 == 0:
                print(f"Iteration {iteration + 1}/{self.max_iter} | Best Loss: {self.best_fitness:.4f}")
        
        return self.best_solution, self.best_fitness
    
    def plot_convergence(self):
        """Plot convergence curve"""
        plt.figure(figsize=(10, 6))
        plt.plot(self.convergence_curve, 'b-', linewidth=2, label='Best Fitness')
        plt.fill_between(range(len(self.convergence_curve)), 
                        self.convergence_curve, alpha=0.3, color='blue')
        plt.xlabel('Iteration')
        plt.ylabel('Validation Loss')
        plt.title('Whale Optimization Convergence Curve')
        plt.grid(True, alpha=0.3)
        plt.legend()
        plt.show()

# ============================================
# GREY WOLF OPTIMIZER
# ============================================

class GreyWolfOptimizer:
    def __init__(self, n_wolves=4, max_iter=8):
        """Grey Wolf Optimizer"""
        self.n_wolves = n_wolves
        self.bounds = self.get_bounds()
        self.max_iter = max_iter
        
        self.alpha_wolf = None
        self.alpha_fitness = float('inf')
        self.beta_wolf = None
        self.beta_fitness = float('inf')
        self.delta_wolf = None
        self.delta_fitness = float('inf')
        
        self.convergence_curve = []
    
    def get_bounds(self):
        """Hyperparameter bounds"""
        return {
            'learning_rate': (1e-4, 1e-2),
            'batch_size': (8, 24),
            'n_filters1': (8, 32),
            'n_filters2': (16, 48),
            'n_filters3': (32, 64),
            'dense_units': (64, 192),
            'dropout_rate': (0.1, 0.4)
        }
    
    def evaluate_wolf(self, wolf_params, train_gen, val_gen):
        """Evaluate wolf parameters"""
        whale_evaluator = WhaleOptimization(n_whales=1, max_iter=1)
        return whale_evaluator.evaluate_whale(wolf_params, train_gen, val_gen)
    
    def optimize(self, train_gen, val_gen):
        """Main GWO optimization loop"""
        print("\n" + "="*60)
        print("GREY WOLF OPTIMIZER")
        print("="*60)
        print(f"Population: {self.n_wolves} wolves")
        print(f"Iterations: {self.max_iter}")
        print("-"*60)
        
        whale_helper = WhaleOptimization(n_whales=self.n_wolves, max_iter=1)
        wolves = whale_helper.initialize_whales()
        fitness = [self.evaluate_wolf(w, train_gen, val_gen) for w in wolves]
        
        sorted_idx = np.argsort(fitness)
        self.alpha_wolf = wolves[sorted_idx[0]].copy()
        self.alpha_fitness = fitness[sorted_idx[0]]
        self.beta_wolf = wolves[sorted_idx[1]].copy()
        self.beta_fitness = fitness[sorted_idx[1]]
        self.delta_wolf = wolves[sorted_idx[2]].copy()
        self.delta_fitness = fitness[sorted_idx[2]]
        
        print(f"Initial alpha loss: {self.alpha_fitness:.4f}")
        
        for iteration in tqdm(range(self.max_iter), desc="GWO Progress"):
            a = 2 - iteration * (2 / self.max_iter)
            
            for i in range(self.n_wolves):
                current_wolf = wolves[i].copy()
                
                for param in self.bounds.keys():
                    r1, r2 = np.random.rand(), np.random.rand()
                    A1 = 2 * a * r1 - a
                    C1 = 2 * r2
                    D_alpha = abs(C1 * self.alpha_wolf[param] - current_wolf[param])
                    X1 = self.alpha_wolf[param] - A1 * D_alpha
                    
                    r1, r2 = np.random.rand(), np.random.rand()
                    A2 = 2 * a * r1 - a
                    C2 = 2 * r2
                    D_beta = abs(C2 * self.beta_wolf[param] - current_wolf[param])
                    X2 = self.beta_wolf[param] - A2 * D_beta
                    
                    r1, r2 = np.random.rand(), np.random.rand()
                    A3 = 2 * a * r1 - a
                    C3 = 2 * r2
                    D_delta = abs(C3 * self.delta_wolf[param] - current_wolf[param])
                    X3 = self.delta_wolf[param] - A3 * D_delta
                    
                    new_val = (X1 + X2 + X3) / 3
                    
                    min_val, max_val = self.bounds[param]
                    if param == 'batch_size':
                        new_val = int(np.clip(new_val, min_val, max_val))
                    else:
                        new_val = np.clip(new_val, min_val, max_val)
                    
                    current_wolf[param] = new_val
                
                new_fitness = self.evaluate_wolf(current_wolf, train_gen, val_gen)
                if new_fitness < fitness[i]:
                    wolves[i] = current_wolf.copy()
                    fitness[i] = new_fitness
            
            sorted_idx = np.argsort(fitness)
            self.alpha_wolf = wolves[sorted_idx[0]].copy()
            self.alpha_fitness = fitness[sorted_idx[0]]
            self.beta_wolf = wolves[sorted_idx[1]].copy()
            self.beta_fitness = fitness[sorted_idx[1]]
            self.delta_wolf = wolves[sorted_idx[2]].copy()
            self.delta_fitness = fitness[sorted_idx[2]]
            
            self.convergence_curve.append(self.alpha_fitness)
            
            if (iteration + 1) % 5 == 0:
                print(f"Iteration {iteration+1}/{self.max_iter} | Alpha Loss: {self.alpha_fitness:.4f}")
        
        return self.alpha_wolf, self.alpha_fitness
    
    def plot_convergence(self):
        """Plot convergence curve"""
        plt.figure(figsize=(10, 6))
        plt.plot(self.convergence_curve, 'r-', linewidth=2, label='Alpha Fitness')
        plt.fill_between(range(len(self.convergence_curve)), 
                        self.convergence_curve, alpha=0.3, color='red')
        plt.xlabel('Iteration')
        plt.ylabel('Validation Loss')
        plt.title('Grey Wolf Optimizer Convergence Curve')
        plt.grid(True, alpha=0.3)
        plt.legend()
        plt.show()

# ============================================
# FINAL MODEL TRAINING
# ============================================

def train_final_model(params, train_gen, val_gen, test_gen, algorithm_name, epochs=30):
    """Train final model with optimized parameters"""
    print(f"\n" + "="*60)
    print(f"TRAINING {algorithm_name.upper()} MODEL")
    print("="*60)
    print(f"Parameters: {params}")
    print(f"Training for {epochs} epochs...")
    print("-"*60)
    
    tf.keras.backend.clear_session()
    gc.collect()
    
    model = keras.Sequential([
        keras.layers.Conv2D(int(params['n_filters1']), (3, 3), 
                           activation='relu', input_shape=(64, 64, 3), padding='same'),
        keras.layers.BatchNormalization(),
        keras.layers.MaxPooling2D((2, 2)),
        keras.layers.Dropout(params['dropout_rate']/3),
        
        keras.layers.Conv2D(int(params['n_filters2']), (3, 3), 
                           activation='relu', padding='same'),
        keras.layers.BatchNormalization(),
        keras.layers.MaxPooling2D((2, 2)),
        keras.layers.Dropout(params['dropout_rate']/2),
        
        keras.layers.Conv2D(int(params['n_filters3']), (3, 3), 
                           activation='relu', padding='same'),
        keras.layers.BatchNormalization(),
        keras.layers.MaxPooling2D((2, 2)),
        keras.layers.Dropout(params['dropout_rate']),
        
        keras.layers.GlobalAveragePooling2D(),
        
        keras.layers.Dense(int(params['dense_units']), activation='relu'),
        keras.layers.Dropout(params['dropout_rate']),
        keras.layers.Dense(train_gen.num_classes, activation='softmax')
    ])
    
    optimizer = keras.optimizers.Adam(learning_rate=params['learning_rate'])
    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    batch_size = min(16, int(params['batch_size']))
    
    callbacks = [
        keras.callbacks.ModelCheckpoint(
            f'{algorithm_name.lower().replace(" ", "_")}_best.keras',
            monitor='val_loss',
            save_best_only=True,
            save_weights_only=False
        ),
        keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=8,
            restore_best_weights=True
        ),
        keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=4,
            min_lr=1e-6
        )
    ]
    
    print(f"Batch size: {batch_size}")
    print(f"Learning rate: {params['learning_rate']:.6f}")
    
    start_time = time.time()
    history = model.fit(
        train_gen,
        steps_per_epoch=train_gen.samples // batch_size,
        validation_data=val_gen,
        validation_steps=val_gen.samples // batch_size,
        epochs=epochs,
        callbacks=callbacks,
        verbose=1
    )
    training_time = time.time() - start_time
    
    test_loss, test_accuracy = model.evaluate(test_gen, verbose=0)
    
    print(f"\n{algorithm_name} RESULTS:")
    print(f"   Test Accuracy: {test_accuracy:.4f}")
    print(f"   Test Loss: {test_loss:.4f}")
    print(f"   Training Time: {training_time/60:.1f} minutes")
    print("="*60)
    
    plt.figure(figsize=(12, 4))
    
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title(f'{algorithm_name} - Accuracy')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(f'{algorithm_name} - Loss')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return model, history, test_accuracy, test_loss

# ============================================
# COMPARISON AND RESULTS
# ============================================

def compare_all_results(baseline_acc, whale_acc, wolf_acc, 
                       baseline_params=None, whale_params=None, wolf_params=None):
    """Compare results of all approaches"""
    print("\n" + "="*60)
    print("COMPARISON OF ALL APPROACHES")
    print("="*60)
    
    print(f"{'Algorithm':<25} {'Test Accuracy':<15} {'Improvement':<15}")
    print("-"*55)
    print(f"{'Baseline':<25} {baseline_acc:<15.4f} {'-':<15}")
    print(f"{'WOA Optimized':<25} {whale_acc:<15.4f} {f'+{(whale_acc - baseline_acc)*100:.2f}%':<15}")
    print(f"{'GWO Optimized':<25} {wolf_acc:<15.4f} {f'+{(wolf_acc - baseline_acc)*100:.2f}%':<15}")
    print("="*60)
    
    algorithms = ['Baseline', 'WOA', 'GWO']
    accuracies = [baseline_acc, whale_acc, wolf_acc]
    
    plt.figure(figsize=(10, 6))
    bars = plt.bar(algorithms, accuracies, color=['blue', 'green', 'red'])
    plt.ylim([0, 1])
    plt.ylabel('Test Accuracy')
    plt.title('Brain Tumor MRI Classification Performance')
    plt.grid(True, alpha=0.3)
    
    for bar, acc in zip(bars, accuracies):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
                f'{acc:.4f}', ha='center', va='bottom', fontweight='bold')
    
    plt.show()
    
    if whale_params and wolf_params:
        print("\nPARAMETER COMPARISON:")
        print("-"*40)
        for param in whale_params.keys():
            print(f"{param:<20} WOA: {whale_params[param]:<10.4f} | GWO: {wolf_params[param]:<10.4f}")

# ============================================
# PRE-TESTED PARAMETERS
# ============================================

def get_pre_tested_parameters():
    """Return pre-tested parameters that work without OOM"""
    return {
        'learning_rate': 0.001,
        'batch_size': 16,
        'n_filters1': 16,
        'n_filters2': 32,
        'n_filters3': 64,
        'dense_units': 128,
        'dropout_rate': 0.3
    }

# ============================================
# EVALUATION FUNCTION
# ============================================

def evaluate_baseline_model(baseline_model, test_generator):
    """Properly evaluate baseline model with multiple metrics"""
    print("\nEvaluating baseline model...")
    
    evaluation_results = baseline_model.evaluate(test_generator, verbose=0)
    
    num_metrics = len(evaluation_results)
    
    if num_metrics == 2:
        baseline_loss, baseline_accuracy = evaluation_results
    elif num_metrics >= 4:
        baseline_loss = evaluation_results[0]
        baseline_accuracy = evaluation_results[1]
        baseline_precision = evaluation_results[2] if num_metrics > 2 else None
        baseline_recall = evaluation_results[3] if num_metrics > 3 else None
    else:
        baseline_loss = evaluation_results[0]
        baseline_accuracy = evaluation_results[1] if num_metrics > 1 else None
    
    print(f"Baseline Test Loss: {baseline_loss:.4f}")
    if baseline_accuracy is not None:
        print(f"Baseline Test Accuracy: {baseline_accuracy:.4f}")
    
    return baseline_loss, baseline_accuracy

# ============================================
# MAIN EXECUTION
# ============================================

def main():
    """Main execution function"""
    print("HYPERPARAMETER OPTIMIZATION FOR BRAIN TUMOR CLASSIFICATION")
    print("="*60)
    print("Using optimized settings to avoid OOM errors")
    print("="*60)
    
    # Load dataset
    print("\nLoading dataset...")
    train_generator, validation_generator, test_generator = load_brain_tumor_dataset()
    
    # Load baseline model
    print("\nLoading/creating baseline model...")
    baseline_model = load_baseline_model()
    
    # Evaluate baseline
    baseline_loss, baseline_accuracy = evaluate_baseline_model(baseline_model, test_generator)
    
    # Ask user for approach
    print("\n" + "="*60)
    print("SELECT OPTIMIZATION APPROACH")
    print("="*60)
    print("1. Use pre-tested parameters (Fast, guaranteed to work)")
    print("2. Run metaheuristic optimization (Slower, may find better params)")
    print("3. Skip optimization, use baseline only")
    
    choice = input("\nEnter choice (1, 2, or 3): ").strip()
    
    if choice == "1":
        print("\nUsing pre-tested parameters...")
        whale_params = get_pre_tested_parameters()
        wolf_params = get_pre_tested_parameters()
        
        wolf_params = wolf_params.copy()
        wolf_params['learning_rate'] = 0.0008
        wolf_params['dropout_rate'] = 0.25
        
    elif choice == "2":
        print("\nRunning metaheuristic optimization...")
        
        print("\n" + "="*60)
        whale_opt = WhaleOptimization(n_whales=7, max_iter=10)
        whale_params, whale_best_loss = whale_opt.optimize(train_generator, validation_generator)
        whale_opt.plot_convergence()
        
        print("\n" + "="*60)
        wolf_opt = GreyWolfOptimizer(n_wolves=7, max_iter=10)
        wolf_params, wolf_best_loss = wolf_opt.optimize(train_generator, validation_generator)
        wolf_opt.plot_convergence()
        
    elif choice == "3":
        print("\nSkipping optimization, using baseline only")
        compare_all_results(baseline_accuracy, baseline_accuracy, baseline_accuracy)
        return
    else:
        print("\nInvalid choice, using pre-tested parameters")
        whale_params = get_pre_tested_parameters()
        wolf_params = get_pre_tested_parameters()
    
    # Train optimized models
    print("\n" + "="*60)
    print("TRAINING OPTIMIZED MODELS")
    print("="*60)
    
    whale_model, whale_history, whale_acc, whale_loss = train_final_model(
        whale_params, train_generator, validation_generator, test_generator,
        "Whale Optimized", epochs=50
    )
    
    wolf_model, wolf_history, wolf_acc, wolf_loss = train_final_model(
        wolf_params, train_generator, validation_generator, test_generator,
        "Grey Wolf Optimized", epochs=50
    )
    
    compare_all_results(baseline_accuracy, whale_acc, wolf_acc, 
                       None, whale_params, wolf_params)
    
    print("\nSaving models...")
    whale_model.save('whale_optimized.keras')
    wolf_model.save('wolf_optimized.keras')
    print("Models saved successfully!")
    
    print("\n" + "="*60)
    print("OPTIMIZATION SUMMARY")
    print("="*60)
    print(f"Baseline Accuracy:     {baseline_accuracy:.4f}")
    print(f"Whale Optimized:       {whale_acc:.4f} ({'+' if whale_acc > baseline_accuracy else ''}{(whale_acc - baseline_accuracy)*100:.2f}%)")
    print(f"Grey Wolf Optimized:   {wolf_acc:.4f} ({'+' if wolf_acc > baseline_accuracy else ''}{(wolf_acc - baseline_accuracy)*100:.2f}%)")
    print(f"Best Algorithm:        {'Whale' if whale_acc > wolf_acc else 'Grey Wolf'}")
    print("="*60)

# ============================================
# RUN THE PROGRAM
# ============================================

if __name__ == "__main__":
    gpus = tf.config.experimental.list_physical_devices('GPU')
    if gpus:
        try:
            for gpu in gpus:
                tf.config.experimental.set_memory_growth(gpu, True)
        except RuntimeError as e:
            print(f"Could not set GPU memory growth: {e}")
    
    main()
```
